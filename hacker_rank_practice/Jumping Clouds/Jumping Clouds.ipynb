#### Jumping Clouds:
<pre>
Emma is playing a game where she must traverse consecutive clouds.
The clouds are either: 1. thunderheads
                       2. cumulus
*Important* - Emma can only traverse to cumulus clouds if their # value is 1 or 2 ints greater than current cloud.

Inputs: n = total # of clouds
        c = clouds = array of binary ints

Example: c = 0,1,0,0,0,1,0
         0 = safe ... 1 = avoid
    Two paths to success here: 1. 0->2->4->6 (3 jumps)
                               2. 0->2->3->4->6 (4 jumps)

Constraints: See document above

Goal: return min # of jumps required to win
      </pre>

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the jumpingOnClouds function below.
def jumpingOnClouds(c):
    
    # Start @ Cloud-0
    currentCloud = 0
    # Keep track of total steps taken
    totalSteps = 0
    # Safe clouds are marked as 0
    safe = 0

    # While current location has not exceeded the len(c)
    while currentCloud in range(n-1):
        # Check if 2 cloud jump <= len(c)
        if (currentCloud+2) <= (n-1):
            # If 2 cloud jump == safe, jump!
            if c[currentCloud+2] == safe:
                currentCloud += 2
                totalSteps += 1
                
        # If 1 cloud jump == safe, jump!
        elif c[currentCloud+1] == safe:
            currentCloud += 1
            totalSteps +=1
                
    # Return total step count
    return totalSteps
             
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    # Number of clouds
    n = int(input())
    
    # List of clouds marked either, 0=safe or 1=avoid
    c = list(map(int, input().rstrip().split()))

    result = jumpingOnClouds(c)

    fptr.write(str(result) + '\n')

    fptr.close()
```
