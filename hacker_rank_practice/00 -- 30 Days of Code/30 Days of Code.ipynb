```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Constraint on input is: N >= 1 and N <= 100 ... 
#    don't worry about outside cases
def conditions():
    # If N is odd
    if N % 2 == 1:
        print('Weird')
    # While N is even
    while N % 2 == 0:
        # If N in range(2,5)
        if N>=2 and N<=5:
            print('Not Weird')
        elif N>=6 and N<=20:
            print('Weird')
        elif N>20:
            print('Not Weird')
        # Break to stop loop
        break


if __name__ == '__main__':
    N = int(input())
    # Return condiitons method
    conditions()
```

## Day 9 -- Recursion 3

<pre>
Practicing an algorithmic concept called Recursion:

Task:
    Write factorial function
        Takes pos int, N as a parameter
            prints result of N!(N factorial)

Given:
    The argument to pass to factorial

Important:
If you fail to use recursion or fail to name your recursive... 
 function factorial or Factorial, you will get a score of 0.

Input Format:
    A single integer, N (arg to pass to factorial)
    
Output Format:
    Print a single int denoting N!

Example Input:
        3
   Example Output:
            6

Constraints: See document above
      </pre>

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the factorial function below.
def factorial(n):
    if n == 0:
        return 1
    else:
        # Not many people know, but python offers a direct function             that can compute the factorial of a number without                  writing the whole code for computing factorial.
        return n * factorial(n - 1)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    result = factorial(n)

    fptr.write(str(result) + '\n')

    fptr.close()
```
