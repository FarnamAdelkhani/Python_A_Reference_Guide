### Day 1 -- Standard Deviation


```python
<pre>
Objective: Calculate standard deviation.
    
Task: Given array C of N ints:
        cal and print the std. dev
            answer should be in decimal form
                rounded to a scale of 1 decimal place
    error margin of +- 0.1 will be tolerated for STD dev
    
Input: N = number of inputs
       N = space-separated 
    
Constraints:
    
Output Format:
    
Sample Input:
    5
    10 40 30 50 20
Sample Output:
    14.1
    
Approach:
numpy.std
numpy.std(a, axis=None, dtype=None, out=None, ddof=0, ...
            keepdims=<no value>)[source]
Doc: https://docs.scipy.org/doc/numpy/reference/generated/numpy.std.html
  
</pre>
```

```python
import math

# Given string of length, L
L = int(input())
string = list(map(int, input().split()))

# µ = mean
µ = sum(string) / L

# Formula for standard deviation
stdev = (sum((i - µ)**2 for i in string) / L) ** 0.5

# Print to 1 decimal place
print("{:.1f}".format(stdev))
```

### Day 1 -- Interquartile Range


<pre>
Objective: Calculate interquartile range.
    
Task: Given an array, X of n ints. F represents freqs of X's.
Calculate: First Quartile (Q1)
           Second Quartile (Q2)
           Third Quartile (Q3)
    
Inputs: 
    
Constraints:
    See document.
    
Output Format:
    
    
Sample Input:
    6
    6 12 8 10 20 16
    5 4 3 2 1 5
    
Sample Output:
    9.0
    
Approach:

  
</pre>

```python
# Define median: even/odd cases
def median(L, values):
    if L % 2 == 0:
        median = (values[int(L/2)-1] + values[int(L/2)]) / 2
    else:
        median = float(values[int(L/2)])
    return int(median)

# Given string of length, L
L = int(input())
ints = list(map(int, input().split()))
freq = list(map(int, input().split()))

new = []
for i in range(len(ints)):
    for j in range(freq[i]):
        new.append(ints[i])
new = sorted(new)

# Determine if the total data is even or odd
L = int(len(new))
if L % 2 == 0:
    low = new[0:int(L/2)]
    high = new[int(L/2):L]
else:
    low = new[0:int(L/2)]
    high = new[int(L/2)+1:L]

# Output data determined by median value
lowest = median(len(low), low)
highest = median(len(high), high)
print("{:.1f}".format(highest-lowest))
```

### Day 2 -- Basic Probability


<pre>If all outcomes are equally likely:
    P(event) = #event_outcomes / #total_outcomes</pre>

```python
import itertools

possible_outcomes = [1, 2, 3, 4, 5, 6]
total_possible_outcomes = list(itertools.product(possible_outcomes, repeat=2))
positive_outcomes = [sum(outcome) for outcome in total_possible_outcomes if sum (outcome)]

probability = len(positive_outcomes)/float(len(total_possible_outcomes))
print(probability)

# Worse case: O(n^k)
```

<pre>Manual method: 
itertools.product:
    Takes dot product of an arbitrary number of iterables. Handy.

Note:
- If both dice role 1 to 4 then guaranteed outcome of < 9
- If a role involves a 5, then other rolls can range(1,4)
- If a role involves a 6, then other rolls can range(1,3)

So favorable cases involve:
    4^2 + 4^1 * 2 + 3^1 * 2 
        = 16 + 8 + 6 
            = 30
            
6^2 = 36 total outcomes possible

... p(sum <= 9) = 30/36 = 5/6</pre>

### Day 2 -- Compound Event Probability


<pre>Probability of a compound event

3 urns: X Y Z

Find:
    Probability that of 3 balls drawn:
        2 are red -and- 1 is black
    
Calculate prob of picking red or black for X, Y, and Z:
    Red  Black

3 Possible picks: (R R B) (R B R) (B R R)
P(R,R,B) = P(R|X)*P(R|Y)*P(B|Z) = 4/7 * 5/9 * 4/8 = 20/126
P(R,B,R) = P(R|X)*P(B|Y)*P(R|Z) = 4/7 * 4/9 * 4/8 = 16/126
P(B,R,R) = P(B|X)*P(R|Y)*P(R|Z) = 3/7 * 5/9 * 4/8 = 15/126

P(#R=2 ∩ #B=1) = P(R,R,B) + P(R,B,R) + P(B,R,R)
P(#R=2 ∩ #B=1) =  20/126  +  16/126  +  15/126 = 51/126 = 17/42

Answer: 17/42
</pre>
