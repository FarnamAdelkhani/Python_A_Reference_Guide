#### Hike:
<pre>
Start @ seaLevel
Each step is 1 unit Δ in altitude, up or down
n = number of steps
s = hike route string = [DDUUUUDD]
Input: U = Step up
       D = Step down
Constraints: See document above

Goal: Find # of valleys walked through *Remember: only concerned about valleys, not mountains.
      Return an int to represent # of valleys, Gary transversed. 
      </pre>

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the countingValleys function below.
def countingValleys(n, s):
    
    # Hike always begins @ seaLevel
    seaLevel = 0
    # Begin: # of valleys traversed
    valleys = 0
    
    # U&D = Δ to seaLevel
    for i in range(n):
        # Up: +1 to seaLevel
        if (s[i] == "U"):
            seaLevel += 1

            # If seaLevel reaches 0: +1 Valleys Traversed
            if seaLevel == 0:
                valleys += 1
                
        # Down: -1 to seaLevel
        elif (s[i] == "D"):
            seaLevel -= 1
            
    # Return total # of valleys        
    return valleys

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    s = input()

    result = countingValleys(n, s)

    fptr.write(str(result) + '\n')

    fptr.close()
```
