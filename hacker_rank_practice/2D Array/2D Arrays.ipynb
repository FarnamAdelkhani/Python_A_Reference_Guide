Find the sum of all values as the hourglass strides through the matrix from top-left to bottom-right.
- There should be x16 sums since the outer ints must be excluded
- Could think of the output as a new matrix, determine the highest value within that new matrix.

Inputs: n = total # of clouds
        c = clouds = array of binary ints

Constraints: See document above

Goal: Find highest output value from convolution of the hourglass matrix across the given matrix named arr.
      </pre>

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Add all ints in Hourglass
def getSumHourglass(matrix, row, col):
    sum = 0
    sum += matrix[row-1][col-1]
    sum += matrix[row-1][col]
    sum += matrix[row-1][col+1]
    sum += matrix[row][col]
    sum += matrix[row+1][col-1]
    sum += matrix[row+1][col]
    sum += matrix[row+1][col+1]
    return sum

def hourglassSum(arr):
    # Begin iteration at Min possible sum (7*(-9)) = -63
    maxSum = -63
    # Avoid outer integers
    for i in range(1,5):
        for j in range(1,5):
            # Extract highest value
            current = getSumHourglass(arr, i, j)
            if current > maxSum:
                maxSum = current

    return maxSum

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    # arr is the outermost array
    arr = []

    # Each line is made into a row array in arr
    for _ in range(6):
        arr.append(list(map(int, input().rstrip().split())))

    result = hourglassSum(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```
