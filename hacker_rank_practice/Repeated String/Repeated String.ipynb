#### Repeated Strings:
<pre>
A string of infinitely repeating chars; find the number of 'a' values in the given range of n.

Input: s = A string input and number of letters to iterate.
       n = Number of chars to iterate through in the given string, s.
Constraints: See document above

Goal: Return the total count of the letter 'a' in the infinetely repeating string s, given the range n.

Hint: Instead of multiplying the length of the string and iterating through it (slow), find the total count of 'a' to plug into an equation.

2nd Hint: Think about decimals and how you'd round your output. Is there a better way?
      </pre>

```python
#!/bin/python3

import math
import os
import random
import re
import sys

def repeatedString(s, n):
    # Quick check
    if len(s) == 1:
        if s == 'a':
            return n
    # Init number of 'a' counted
    aCount = 0
    # Iterate string, search and count 'a'
    for i in s:
        if i == 'a':
            aCount += 1
    
    # Instead of iterating the entire string
    # Find # of 'a' in s, then use an equation 
    total = aCount * math.floor(n / len(s))
    # Total was floored in last step, now count remaining 'a'
    total += s[:(n%len(s))].count('a')
    
    return total

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    # String input
    s = input()

    # Number of characters to iterate
    n = int(input())

    result = repeatedString(s, n)

    fptr.write(str(result) + '\n')

    fptr.close()
```
